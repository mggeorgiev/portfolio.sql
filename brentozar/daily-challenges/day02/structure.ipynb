```sql
/* To solve we first create a table with all possbile combination.
   Three additional columns are added to store the value and the outcome of the round.
*/

USE tempdb;
GO

IF OBJECT_ID('[dbo].[day_two_perm]', 'U') IS NOT NULL
DROP TABLE [dbo].[day_two_perm]
GO
/* Create the table in the specified schema */
CREATE TABLE [dbo].[day_two_perm]
(
    [Id] INT identity(1,1), /* Primary Key column */
    [move_one] varchar(1),
    [move_two] varchar(1),
    [move_one_value] int,
    [move_two_value] int,
    [result] int
);
GO

INSERT INTO [dbo].[day_two_perm]([move_one],[move_two] )
    SELECT
        * 
        FROM (VALUES ('A'), ('B'), ('C')) as t1(one)
        CROSS JOIN (VALUES ('X'),('Y'), ('Z')) as t2(two);
```

```sql
/* Time to update the values.*/ 
USE tempdb;
GO

UPDATE [dbo].[day_two_perm]
    SET [move_one_value] = CASE
                                WHEN move_one = 'A' THEN 1
                                WHEN move_one = 'B' THEN 2
                                WHEN move_one = 'C' THEN 3
                           END;

UPDATE [dbo].[day_two_perm]
    SET [move_two_value] = CASE
                                WHEN move_two = 'X' THEN 1
                                WHEN move_two = 'Y' THEN 2
                                WHEN move_two = 'Z' THEN 3
                           END;

UPDATE [dbo].[day_two_perm]
    SET [result] = move_two_value + CASE
                                        WHEN move_one_value > move_two_value THEN 0
                                        WHEN move_one_value = move_two_value THEN 3
                                        WHEN move_one_value < move_two_value THEN 6
                                    END;

/*
Since there are only 3 posible outcomes we could also update in sequence:

UPDATE [dbo].[day_two_perm]
     SET [result] = move_two
     WHERE move_one_value > move_two_value;

UPDATE [dbo].[day_two_perm]
     SET [result] = move_two + 3
     WHERE move_one_value = move_two_value;

UPDATE [dbo].[day_two_perm]
     SET [result] = move_two_value + 6
     WHERE move_one_value < move_two_value;
*/

SELECT * FROM [dbo].[day_two_perm];
```
